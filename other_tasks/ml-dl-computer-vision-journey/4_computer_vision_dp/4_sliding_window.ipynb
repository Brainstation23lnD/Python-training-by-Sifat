```python
from tensorflow.keras import layers
from tensorflow_hub.keras_layer import keras

model = keras.Sequential([
    layers.Conv2D(
        64, (3, 3),
        activation='relu',
        padding='same',
    ),
    layers.MaxPooling2D(
        (2, 2),
        strides=1,
        padding='same',
    ),
])
```

```python
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# Dummy input image: 32x32, 3 channels
random_state = 42
rng = np.random.default_rng(random_state)

x = rng.random((1, 32, 32, 3), dtype=np.float32)

# Conv2D layer with 64 filters
conv_layer = layers.Conv2D(
    64, (3, 3),
    activation='relu',
    padding='same'
)

# Forward pass
feature_maps = conv_layer(x)  # shape: (1, 32, 32, 64)
print("Feature maps shape:", feature_maps.shape)

# Visualize first 10 feature maps
plt.figure(figsize=(15, 3))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(feature_maps[0, :, :, i], cmap='gray')
    plt.axis('off')
    plt.title(f'Filter {i+1}')
plt.show()
```
